## One hot encoding frequent categories

```python
# we can turn the previous commands into 2 functions


def calculate_top_categories(df, variable, how_many=5):
    return [
        x for x in df[variable].value_counts().sort_values(
            ascending=False).head(how_many).index
    ]


def one_hot_encode(data, variable, top_x_labels):

    for label in top_x_labels:
        data[variable + '_' + str(label)] = np.where(
            data[variable] == label, 1, 0)
    
```

```python
data_onehot_freq=data[var_max_3catf]
```

```python
for variable in var_max_3catf:
    
    top_categories = calculate_top_categories(data_onehot_freq, variable, how_many=5)
    
    one_hot_encode(data_onehot_freq, variable, top_categories)
```
