## Distribucion de los data missing

### valores sin nan

```python
vars_with_sin_na = [var for var in data_features.columns if data_features[var].isnull().mean() ==0]
```

### valores con nan de 0% a 5%

```python
vars_with_title_na = [var for var in data_features.columns if data_features[var].isnull().mean() >0 and data_features[var].isnull().mean() <=0.05]
```

### valores con nan de 6% a 30%

```python
vars_with_median_na = [var for var in data_features.columns if data_features[var].isnull().mean() >0.05 and data_features[var].isnull().mean() <=0.30]
```

### valores con nan de 31% a 60%

```python
# Cantidad de variables que eliminare por que tienen mas del 50% de valores faltantes
vars_with_big_na = [var for var in data_features.columns if data_features[var].isnull().mean() >0.30 and data_features[var].isnull().mean() <=0.60]
```

### valores con nan de mayor 60% y menor a 80%

```python
vars_with_very_big_na = [var for var in data_features.columns if data_features[var].isnull().mean() > 0.60 and data_features[var].isnull().mean() <= 0.80 ]
```

### valores con nan de mayor 80%

```python
vars_with_ultra_big_na = [var for var in data_features.columns if data_features[var].isnull().mean() > 0.80 ]
```

### Distribucion total  de valores faltantes

```python
print(f"cantinad de valores que tienen valores nan : {len(vars_with_sin_na)}")
print(f"cantinad de valores que tienen entre 1% y 15% de valores nan : {len(vars_with_title_na)}")
print(f"cantinad de valores que tienen entre 16% y 30% de valores nan : {len(vars_with_median_na )}")
print(f"cantinad de valores que tienen entre 31% y 60% de valores nan : {len(vars_with_big_na)}")
print(f"cantinad de valores que tienen mas del 61% de valores nan: {len(vars_with_very_big_na)}")
print(f"cantidad de variables totales: {len(data_features.columns)}")
```
